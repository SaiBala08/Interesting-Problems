```python
class SoftwareComapnies:
    
    def __init__(self, companyNames, processList, cost, amount, Company1, Company2):
        
        self.companyNames = companyNames
        self.processList = processList
        self.cost = cost
        self.amount = amount
        self.company1 = Company1
        self.company2 = Company2
        
    
    def produceData(self):
        
        C_Dic = {}
        C_list = []
        return_list = [self.company1]
        i = self.companyNames.index(self.company1)
        process_i_data = self.processList[i].split(" ")

        if self.company2 in process_i_data:
            return_list.append(self.company2)
            return_list.sort()
            return return_list

        for a in process_i_data:
            if self.processList[self.companyNames.index(a)] != "":
                C_list.append(a)
                C_Dic[a] = self.amount[self.companyNames.index(a)]
                
         
        return_list += C_list
        check_list = return_list.copy()

        print("C_list",C_list)
        print("return_list",return_list)
        print("check_list",check_list)
        print("\n")


        while(C_list):
            
            print("C_list",C_list)
            current_company = C_list.pop()
            i = self.companyNames.index(current_company)
            
            print("C_list",C_list)

            current_cost = self.cost[i]
            current_amount = self.amount[i]

            print("current_com-",current_company,"Current com data-",self.amount[i])
            
            
            if " " in self.processList[i]:
                data_i_companies = self.processList[i].split(" ")
            else:
                data_i_companies = [self.processList[i]]
                if data_i_companies == [self.company2]:
                    if self.company2 not in return_list:
                        return_list.append(self.company2)
                    print("Iteration Complete")
                    print("\n")
                    continue

            print("data_i_companies ",data_i_companies)
            print("\n")
            
            temp_Com_list = []
            temp_cost_list = []
            temp_amount_list = []

            for a in data_i_companies:

                if self.processList[self.companyNames.index(a)] != "":    
                    temp_Com_list.append(a)
                    temp_cost_list.append(self.cost[self.companyNames.index(a)])
                    temp_amount_list.append(self.amount[self.companyNames.index(a)])
                    #C_list.append(a)
                    


            print("temp_Com_list",temp_Com_list) 

            temp_Com_list = [x for _,x in sorted(zip(temp_cost_list,temp_Com_list))]
            print("temp_Com_list",temp_Com_list) 
            temp_Com_list.reverse()

            while(True):
                item = temp_Com_list.pop()
                current_amount -= self.amount[self.companyNames.index(item)]
                C_list.append(item)
                if item not in return_list:
                    return_list.append(item)
                if temp_Com_list == [] or current_amount <0:
                    break
            
            if return_list == check_list:
                return []
            
            print("\n")
            print("temp_Com_list",temp_Com_list) 
            print("temp_cost_list",temp_cost_list)
            print("temp_amount_list",temp_amount_list)
            print("C_list",C_list)
            print("return_list",return_list)
            print("Iteration Complete")
            print("\n")
        
        return_list.sort()
        print(return_list)
        return return_list
    
    
```

```python
C = SoftwareComapnies(["topcoder", "doodle", "nasa", "ninny", "idm", "noname", "kintel"],["doodle nasa noname", "idm ninny noname", "idm ninny noname", "kintel", "kintel", "", ""],[1, 2, 7, 4, 6, 1, 2],[50, 10, 11, 9, 14, 11, 23],"topcoder","kintel")
assert C.produceData() == ["doodle", "idm", "kintel", "nasa", "ninny", "topcoder"] , "Error"
```
